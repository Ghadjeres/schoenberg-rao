### Rao Quadradic Entropy on a metric space with two points

```python
rs = np.linspace(0, 10, 200)
thetas = np.linspace(0, 1, 200)

res = np.zeros((len(rs), len(thetas)))
for i, r_val in enumerate(rs):
    for j, theta_val in enumerate(thetas):
        K = torch.tensor([[0, r_val],
                          [r_val, 0]])
        p = torch.tensor([[theta_val, 1-theta_val]])
        res[i, j] = SR.rao_sim_entropy(K, p)
        
        
fig = plt.figure(figsize=(3, 3))
s = fig.add_subplot(1, 1, 1, xlabel='$\\theta$ %', ylabel='$d$')
im = s.imshow(res, origin='lower', cmap=cm.coolwarm)
plt.grid(False)
fig.colorbar(im)
plt.show()

fig = plt.figure(figsize=(3.4, 3.4))
ax = fig.gca(projection='3d')
ax.view_init(30, -235)
ax.set_yticks([0, 2.5, 5])
ax.set_ylabel('$r$', fontsize=12)
ax.set_xticks([0, 0.5, 1])
ax.set_xlabel('$\\theta$', fontsize=12)
ax.grid(False)
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
X, Y = np.meshgrid(thetas, rs)
surf = ax.plot_surface(X, Y, res, cmap=cm.coolwarm, linewidth=3, antialiased=True, alpha=0.8)
cbaxes = fig.add_axes([0.05, 0.15, 0.03, 0.6])  # This is the position for the colorbar
cb = plt.colorbar(surf, cax=cbaxes, shrink=0.8)
plt.savefig("rao_ent_2d", dpi=300, facecolor=None)
#fig.colorbar(surf, shrink=1, orientation='horizontal', pad=0.1, aspect=20)
ax.dist = 11
plt.show()
```

### Finding distribution with maximum Rao's quadratic entropy

```python
K = torch.Tensor(np.array([[0, 3, 4], [3, 0, 5], [4, 5, 0]]))
logit_q = torch.nn.Parameter(torch.Tensor(np.random.randn(1, 3)))
optimizer = torch.optim.Adam([logit_q], lr=0.005, amsgrad=True)
```

```python
q_hist = []
for _ in range(1000):
    optimizer.zero_grad()
    q = torch.softmax(logit_q, dim=1)
    loss = - SR.rao_sim_entropy(K, q)
    
    if _ % 200 == 0:
        print("%d - %.4e" % (_, -loss.item()))
    if _ % 20 == 0:
        q_hist.append(q[0].data.numpy())
    
    loss.backward()
    optimizer.step()

print(torch.softmax(logit_q, dim=1))
```

```python
# Evaluate entropy over simplex to plot heatmap
D = []
for i, theta in enumerate(np.linspace(0.0001, 1-0.0001, 50)):
    for j, rho in enumerate(np.linspace(0.0001, 1-0.0001, 50)):
        if theta + rho <= 1:
            mu = np.array([[theta], [rho], [1-theta-rho]])
            D.append([theta, rho, 1-theta-rho, SR.rao_sim_entropy(K, torch.Tensor(mu.T))])
D = np.array(D)

# translate the data to cartesian corrds
x = D[:, 0] * 4
y = D[:, 1] * 3

# create a triangulation out of these points
T = Triangulation(x,y)

# plot the contour
plt.figure(figsize=(8, 5))
plt.tricontourf(x, y, T.triangles, D[:, -1], 15, cmap=cm.coolwarm, antialiased=True, alpha=0.9)
#plt.colorbar()
# 
# # Display Shannon's optimum
x_star, y_star = 1/3 * 4,  1/3 * 3
plt.scatter(x_star, y_star, marker='h', color='w', s=50)

# Display RAO's optimum
#mu_star = q[0].data.numpy()
marker = '+'
for mu_star in q_hist: 
    x_star, y_star = mu_star[0]*4,  mu_star[1]*3
    
    plt.scatter(x_star, y_star, marker=marker, color='k', s=3)
    marker = '.'
plt.scatter(x_star, y_star, marker='*', color='k', s=70)
    

plt.text(0 - 0.1, 0 - 0.2, "A", fontsize=12)
plt.text(0- 0.1, 3 + 0.1, "B", fontsize=12)
plt.text(4 - 0.2, 0 - 0.2, "C", fontsize=12)
plt.axis('scaled')


plt.savefig("toy_heatmap_rao.pdf", dpi=600, facecolor=None)
plt.show()
```

```python
# test minimization

# minimization between two random Normal distributions
mu1 = torch.randn(1)
sigma1 = torch.rand(1)
mu2_init = torch.randn(1)
log_sigma2_init = torch.randn(1)

# minimization between two fixed Normal distributions
# mu1 = torch.ones(1)
# sigma1 = torch.ones(1)
# mu2_init = -torch.ones(1)
# log_sigma2_init = torch.zeros(1)


# SR
print('SR')
mu2 = torch.nn.Parameter(mu2_init.detach())
log_sigma2 = torch.nn.Parameter(log_sigma2_init.detach())

optimizer = torch.optim.Adam([mu2, log_sigma2], lr=0.001)
hist_SR = []
for i in range(5000):
    optimizer.zero_grad()
    sigma2 = torch.exp(log_sigma2)
    loss = SR.SR_gaussian(mu1, sigma1, mu2, sigma2)
    loss.backward()
    optimizer.step()
    if i % 100 == 0: 
        hist_SR.append([mu2.detach().item(), sigma2.detach().item()])
    if i % 1000 == 0: 
        print(loss.item())
print(f'Reference: N({mu1}, {sigma1}')
print(f'Result: N({mu2}, {sigma2}')

# KL
print('KL')
mu2_init = -torch.ones(1)
log_sigma2_init = torch.zeros(1)
mu2 = torch.nn.Parameter(mu2_init)
log_sigma2 = torch.nn.Parameter(log_sigma2_init)
optimizer = torch.optim.Adam([mu2, log_sigma2], lr=0.001)
hist_KL = []
def KL(mu1, sigma1, mu2, sigma2):
    return 1 / 2 *( ( sigma1 ** 2 / sigma2 ** 2) + (mu2 - mu1)**2 / sigma2 **2 - 1 + 2 * torch.log(sigma2 / sigma1))

for i in range(5000):
    optimizer.zero_grad()
    sigma2 = torch.exp(log_sigma2)
    loss = KL(mu1, sigma1, mu2, sigma2)
    loss.backward()
    optimizer.step()
    if i % 100 == 0: 
        hist_KL.append([mu2.detach().item(), sigma2.detach().item()])
    if i % 1000 == 0: 
        print(loss.item())
print(f'Reference: N({mu1}, {sigma1}')
print(f'Result: N({mu2}, {sigma2}')

# W2
print('W2')
mu2_init = -torch.ones(1)
log_sigma2_init = torch.zeros(1)
mu2 = torch.nn.Parameter(mu2_init)
log_sigma2 = torch.nn.Parameter(log_sigma2_init)
optimizer = torch.optim.Adam([mu2, log_sigma2], lr=0.001)
hist_W2 = []
def W2(mu1, sigma1, mu2, sigma2):
    return (mu1 - mu2)**2 + (sigma1 - sigma2)**2

for i in range(5000):
    optimizer.zero_grad()
    sigma2 = torch.exp(log_sigma2)
    loss = W2(mu1, sigma1, mu2, sigma2)
    loss.backward()
    optimizer.step()
    if i % 100 == 0: 
        hist_W2.append([mu2.detach().item(), sigma2.detach().item()])
    if i % 1000 == 0:
        print(loss.item())
print(f'Reference: N({mu1}, {sigma1}')
print(f'Result: N({mu2}, {sigma2}')
```

### Check that SR is metric

```python
# Check that SR is metric (triangle inequality)
num_atoms = 1
for i in range(10000):
    mu_1 = torch.randn(num_atoms) * 1
    sigma_1 = torch.rand(num_atoms) * 1
    mu_2 = torch.randn(num_atoms) * 1
    sigma_2 = torch.rand(num_atoms) * 1
    mu_3 = torch.randn(num_atoms) * 1
    sigma_3 = torch.rand(num_atoms) * 1
    d_12 = torch.sqrt(SR.SR_gaussian(mu_1, sigma_1, mu_2, sigma_2))
    d_13 = torch.sqrt(SR.SR_gaussian(mu_1, sigma_1, mu_3, sigma_3))
    d_23 = torch.sqrt(SR.SR_gaussian(mu_3, sigma_3, mu_2, sigma_2))
    if d_12 + d_13 < d_23:
        print('23')
    if d_12 + d_23 < d_13:
        print('13')
    if d_23 + d_13 < d_12:
        print('12')
```

### Multivariate Gaussian mixtures simplification using Sliced Schoenberg-Rao

```python
# minimization multivariate diagonal
def eval_diag_gaussian(x, mu, sigma):
    d = mu.size(-1)
    exponent = ((x - mu)**2 / sigma)
    denom = (np.pi * 2)**(d / 2) * torch.sqrt(torch.prod(sigma))
    return torch.exp(-1 / 2 * exponent) / denom

def random_diag_gaussian(d):
    mu = torch.randn(d)
    sigma = torch.rand(d) * 2
    return mu, sigma

def random_diag_mixture(d, num_components):
    pi = torch.randn(num_components)
    pi = torch.softmax(pi, dim=0)
    c = [random_diag_gaussian(d) for _ in range(num_components)]
    return pi, c

def SR_diag_mixtures(m1, m2, d):
    # project
    b = torch.randn(d)
    b /= torch.norm(b, p=2)
    pi1, c1 = m1
    pi2, c2 = m2
    
    mus1 = torch.cat([a.unsqueeze(0) for (a,b) in c1], dim=0)
    mus2 = torch.cat([a.unsqueeze(0) for (a,b) in c2], dim=0)
    sigmas1 = torch.cat([b.unsqueeze(0) for (a,b) in c1], dim=0)
    sigmas2 = torch.cat([b.unsqueeze(0) for (a,b) in c2], dim=0)
    
    mu1 = (b*mus1).sum(1)
    mu2 = (b*mus2).sum(1)
    
    sigma1 = torch.sqrt((b * b * sigmas1 * sigmas1).sum(1))
    sigma2 = torch.sqrt((b * b * sigmas2 * sigmas2).sum(1))
    m1_proj = list(zip(pi1, mu1, sigma1))
    m2_proj = list(zip(pi2, mu2, sigma2))
    return SR.SR_mixtures(m1_proj, m2_proj)
```

```python
d = 2
num_components_target = 5
num_components = 3
target = random_diag_mixture(d, num_components_target)

# to param
log_pi = torch.nn.Parameter(torch.randn(num_components))
mu = torch.nn.Parameter(torch.randn(num_components, d))
log_sigma = torch.nn.Parameter(torch.randn(num_components, d))

optimizer = torch.optim.Adam([log_pi, mu, log_sigma], lr=1e-2)

print(target)
for i in range(5000):
    optimizer.zero_grad()
    pi = torch.softmax(log_pi, dim=0)
    sigma = torch.exp(log_sigma)
    
    m = pi, [(a, b) for (a,b) in zip(mu, sigma)]
    loss = SR_diag_mixtures(m, target, d)
    loss.backward()
    optimizer.step()
    if i % 1000 == 0:
        print(loss)
print(m)

```

```python
# plot diag
import seaborn as sns
def eval_diag_gaussian(x, mu, sigma):
    d = mu.size(-1)
    exponent = ((x - mu)**2 / sigma ** 2).sum()
    denom = (np.pi * 2)**(d / 2) * torch.prod(sigma)
    return torch.exp(-1 / 2 * exponent) / denom

def eval_diag_mixture(x, m):
    sum = 0
    for pi, (mu, sigma) in zip(m[0], m[1]):
        sum += pi * eval_diag_gaussian(x, mu, sigma)
    return sum

fig = plt.figure(figsize=(15,10))
lims = (-3.5, 4)
def plot_diag_mixture(m, fig, index, lims=lims):
    X, Y = [], []
    Z = np.zeros((100, 100))
    a, b = lims

    for i, x in enumerate(np.linspace(a, b, 100)):
        for j, y in enumerate(np.linspace(a, b, 100)):
            z = eval_diag_mixture(torch.Tensor([x, y]), m)
            Z[i,j] = z.item()

    X = np.array(np.linspace(a, b, 100))
    Y = np.array(np.linspace(a, b, 100))
    Z = np.array(Z)
    ax = fig.add_subplot(index)
    ax.set_xlim(a,b),  ax.set_ylim(a, b)
    plt.contour(X, Y, Z)
    

# plot simplification and target
plot_diag_mixture(m, fig, 121)
plot_diag_mixture(target, fig, 122)

plt.savefig('multivariate_mixtures.pdf')

    
```
